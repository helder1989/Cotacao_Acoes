## Introdução


**Cotação** é o ato de fixar um valor, precificar elementos e artigos. Ela serve tanto para o câmbio, como para outros ativos, como produtos agrícolas e pecuários ou produtos **financeiros**. 

Em suma **Cotação** é o preço de uma ação, ou seja, quanto o investidor deve pagar para adquiri-la ou quanto receberia pela sua venda. Ela é sempre expressa pelo valor unitário da ação.

O termo **Cotação** se refere à definição de preço de um ativo em um momento específico. Via de regra, ele depende de oferta e demanda, mas também sofre interferência de outros fatores. Na prática, se a procura aumenta, o preço também se eleva. Por outro lado, se a oferta for maior, a cotação despenca.

O experimento tem por objetivo falar sobre o mercardo financeiro, no sentido de apresentar o caminho para extrair cotações de ações automaticamente, tanto de uma ação quanto de uma carteira de ações.

A biblioteca principal será á **pandas_datareader** que vai nos conceder o mecanismo de puxar a cotação de vários lugares diferentes, um deles é o **Yahoo Finance** que é uma das melhores referências para buscar cotações quando o assunto é **programação**. ## Extraindo cotação de 1(uma) ação-Ibovespa

```python
from pandas_datareader import data as web #não vamos usar todos os recursos da biblioteca, só será utilizado 
##a função **data**, sendo que pra não confudir vamos chamar o data de web

cotacao_bovespa = web.DataReader('^BVSP', data_source='yahoo', start="01/01/2020",end="01/01/2021")

display(cotacao_bovespa)
```

**Ticker=** Obtido através do Yahoo Finance (https://finance.yahoo.com/), que é **^BVSP**, ou seja, para buscar cotação de ações de outras instituições ou empresa, basta buscar na lupa que fica no centro do site.

## Vamos usar com referência a coluna "Adj Close"(fechamento ajustado) para realizar a cotação 


```python
##Exibindo essa informação em um gráfico

cotacao_bovespa = web.DataReader('^BVSP', data_source='yahoo', start="01/01/2020",end="01/01/2021")
display(cotacao_bovespa)
cotacao_bovespa["Adj Close"].plot(figsize=(12, 10)) #criando o nosso gráfico inclusive ajustando o tamanho
plt.show() #mostrando o gráfico criado
```

O gráfico nos apresenta as cotações desde a data inicial(01/01/2020 até 01/01/2021). Inclusive se na busca não for encontrado por exemplo a cotação do dia 01/01/2020 (que é feriado) será extraído a cotação do próximo dia. Sendo assim o **start** tenta começar pela data estabelecida, porém não sendo possível vai extrair a cotação do dia seguinte. Já o **end** se o mesmo não conseguir finalizar na data mencionada ele encerra no dia anterior, no nosso exemplo seria em 31/12/2020.

## Extraindo cotação de uma carteira de ações em um intervalo de datas específicas 

Temos um arquivo em excel no qual consta algumas empresas, no qual vamos realizar a cotações de diversas ações.

```python
tabela_empresas = pd.read_excel("Empresas.xlsx") #ler o arquivo em excel
display(tabela_empresas)
```

Lista com ticker de algumas empresas: 

ITUB4: Itaú,
PETR4: Petrobras,
BBDC4: Bradesco,
VALE3: Vale.

```python
tabela_empresas = pd.read_excel("Empresas.xlsx") #ler o arquivo em excel
display(tabela_empresas)

for empresa in tabela_empresas['Empresas']:
    print(empresa) #vai mostrar a tabela e o gráfico de forma organizada
    cotacao = web.DataReader(f'{empresa}.SA', data_source='yahoo', start="01/01/2020",end="01/01/2021")
    display(cotacao)
    cotacao["Adj Close"].plot(figsize=(12, 10)) 
    plt.show() 
```

O **for** está percorrendo um conjunto de informações e armazenando na variável **empresa**,ou seja, a cada vez que rodamos o nosso código ele armazena dentro da variável **empresa** o valor. Sendo assim a primeira vez que ele roda a variável empresa será ITUB4, a segunda vez será PETR4 e vai seguindo até chegar no final da lista. 
Já o **f** na frente do texto serve para indicar que a string 'tamanho do playlist': será concatenada. Sendo uma string que será formatada. Vamos colocar um valor de uma variavel dentro da string para que seja substituída de forma dinâmica
